```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
r = np.linspace(0,20000,1000)
f1 = 10**(0.1*r/5000)
f2 = 10**(0.5*r/5000)
f3 = 10**(1*r/5000)
f4 = 10**(1.5*r/5000)
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
plt.plot(r, np.log10(f1), 'r', label = 'Av = 0.1')
plt.plot(r, np.log10(f2), 'g', label = 'Av = 0.5')
plt.plot(r, np.log10(f3), 'b', label = 'Av = 1.0')
plt.plot(r, np.log10(f4), 'k', label = 'Av = 1.5')
plt.xlabel('Distance (parsecs)')
plt.ylabel('Log of apparent distance to true distance ratio')
plt.title('Log of apparent distence to true distance ratio as afunction of true distance')
plt.legend()
plt.savefig('distancesgalaxy3.png')
```

```python
p1 = 10**(-2*0.1*r/5000)
p2 = 10**(-2*0.5*r/5000)
p3 = 10**(-2*1*r/5000)
p4 = 10**(-2*1.5*r/5000)
```

```python
plt.clf()
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
plt.plot(r, np.log10(p1), 'r', label = 'Av = 0.1')
plt.plot(r, np.log10(p2), 'g', label = 'Av = 0.5')
plt.plot(r, np.log10(p3), 'b', label = 'Av = 1.0')
plt.plot(r, np.log10(p4), 'k', label = 'Av = 1.5')
plt.xlabel('Distance (parsecs)')
plt.ylabel('Log of Apparent density to true density ratio')
plt.legend()
plt.title('Log of Apparent density to true density ratio as a function of distance')
plt.savefig('densitiesgalaxy3.png')
```
