```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal, random
```

```python
mag1 = np.random.normal(0, 0.1, 10000)
mag2 = np.random.normal(0, 0.3, 10000)
mag3 = np.random.normal(0, 0.5, 10000)
mag4 = np.random.normal(0, 1.0, 10000)
```

```python
r = 20000*(np.random.uniform(0, 1, 10000))**(1/3)
print(r)
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
plt.hist(mag1,bins = [-.4, -.35, -.3, -.25, -.2, -.15, -.1, -.05, 0, .05, .1, .15, .2, .25, .3, .35, .4])
plt.xlabel('Absolute Magnitude')
plt.ylabel('Counts')
plt.title('Counts binned at .05 magnitude for .1 magnitude dispersion')
plt.savefig('histogram.png')
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
rnorm = np.linspace(0,20000,1000)
dN = 4*np.pi*(rnorm)**2*2.984*10**-10*100
plt.hist(r, np.arange(0, 20100, 100), color = 'b', label = 'Counts')
plt.plot(rnorm,dN,'--r', label = 'Predicted N')
plt.yscale('log')
plt.xscale('log')
plt.ylim(.001,2*10**2)
#plt.xlim(.001, 20000)
plt.xlabel('Log10(r) (parsecs)')
plt.ylabel('Log10(N) (counts)')
plt.legend()
plt.title('Log10(N) (counts) vs Log10(r) (parsecs)')
plt.savefig('starCounts.png')
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
mod = 5*np.log10(r/10)
amag1 = mod +mag1
ii = amag1 <=14
iii = amag1 <= 30
plt.hist(amag1[ii], color = 'r')#, np.arange(-9, 7.5, .5))
plt.xlabel('Apparent magnitude')
plt.ylabel('Counts')
plt.title('Counts binned at different apparent magnitudes with cutoff magnitude of 14')
plt.savefig('brightcutoff01mag.png')
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
plt.hist(amag1[iii], color = 'k')#, np.arange(-9, 13.5, .5))
plt.xlabel('Apparent magnitude')
plt.ylabel('Counts')
plt.title('Counts binned at different apparent magnitudes with cutoff magnitude of 17')
plt.savefig('faintcutoff01mag.png')
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
plt.plot(r[iii],mag1[iii],'b+', label = 'mag < 17')
plt.plot(r[ii],mag1[ii],'r+', label = 'mag<14')
plt.xlabel('Distance (pc)')
plt.xlim(0,12500)
plt.ylabel('Absolute magnitude')
plt.title('Absolute magnitude as a function of distance for the two selections, 0.1 mag distribution')
plt.legend()
plt.savefig('selectionsmag01mag.png')
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
amag4 = mod +mag4
ii4 = amag4 <=14
iii4 = amag4 <= 20
plt.hist(amag4[ii4], color = 'g')#, np.arange(-9, 6.5, .5))
plt.xlabel('Apparent magnitude')
plt.ylabel('Counts')
plt.title('Counts binned at different apparent magnitudes with cutoff magnitude of 14')
plt.savefig('brightcutoff1mag.png')
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
plt.hist(amag4[iii4], color = 'b')#, np.arange(-9, 13.5, .5))
plt.xlabel('Apparent magnitude')
plt.ylabel('Counts')
plt.title('Counts binned at different apparent magnitudes with cutoff magnitude of 20')
plt.savefig('faintcutoff1mag.png')
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
plt.plot(r[iii4],mag4[iii4],'b+', label = 'mag < 20')
plt.plot(r[ii4],mag4[ii4],'r+', label = 'mag<14')
plt.xlabel('Distance (pc)')
plt.ylabel('Absolute magnitude')
plt.title('Absolute magnitude as a function of distance for the two selections, 1 mag distribution')
plt.legend()
plt.savefig('selectionsmag1mag.png')
```

```python
mest4 = np.random.normal(mag4, np.abs(0.1*mag4), 10000)
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
plt.plot(r[ii4],mag4[ii4]-mest4[ii4],'.b')
plt.xlabel('Distance (parsecs)')
plt.ylabel('Mactual - Mest')
plt.title('Mactual - Mest as a function of distance')
plt.savefig('mest.png')
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
plt.plot(r[ii4],mag4[ii4] -0, '.g')
plt.xlabel('Distance (parsecs)')
plt.ylabel('Mactual - <M>')
plt.title('Mactual - <M> as a function of distance')
plt.savefig('mavg.png')
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
plt.plot(r[ii4], (amag4[ii4]-mest4[ii4])-(amag4[ii4]-mag4[ii4]),'.r')
plt.xlabel('Distance (parsecs)')
plt.ylabel('(m-Mest)-(m-Mactual)')
plt.title('Mest distance modulus minus Mactual distance modulus')
plt.savefig('apparent-absolute.png')
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
plt.plot(r[ii4], (amag4[ii4]-0)-(amag4[ii4]-mag4[ii4]),'.b')
plt.xlabel('Distance (parsecs)')
plt.ylabel('(m-<M>)-(m-Mactual)')
plt.title('<M> distance modulus minus Mactual distance modulus')
plt.savefig('apparent-absolutebias.png')
```
