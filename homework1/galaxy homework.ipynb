```python
plt.clf()
M = 3.09*10**25*xnew*intavg*intavg/((2*10**30)*6.67*(10**(-11)))
pfit = np.polyfit(np.log10(xnew),np.log10(M),1)
p = np.poly1d(pfit)
plt.plot(np.log10(xnew),np.log10(M),'.r',np.log10(xnew),p(np.log10(xnew)),'-k')
plt.xlabel('Log10(R) (kpc)')
plt.ylabel('Log10(M(R)) (Solar Masses)')
plt.savefig("mass.png")
```
